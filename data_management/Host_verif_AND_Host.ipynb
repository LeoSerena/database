```python
barcelona_listings = barcelona_listings[["host_id", "host_verifications"]]
madrid_listings = madrid_listings[["host_id", "host_verifications"]]
berlin_listings = berlin_listings[["host_id", "host_verifications"]]
tab = pd.concat([barcelona_listings, madrid_listings,berlin_listings]).drop_duplicates()
tab.head()
```

```python
tab['host_verifications'] = tab['host_verifications'].apply(lambda x:x.replace('[','').replace(']','').replace("'",''))
list = tab['host_verifications'].str.split(',')#, expand = True)
tab['host_verifications'] = list
tab
```

```python
#print(type(tab['host_verifications'].iloc[0]))
print((tab['host_verifications'].iloc[0][0]))
```

```python
d = host_verificaions.to_dict()['HOST_VERIFICATION']
inv_d = dict([[v,k] for k,v in d.items()])
print(inv_d)
```

```python
tab['host_verifications'] = tab['host_verifications'].apply(lambda am_list : [inv_d[x] for x in am_list if len(am_list)>1])
print(tab)
```

```python
tab = pd.DataFrame(tab.host_verifications.tolist(), index=tab.host_id).stack().reset_index(level=1, drop=True).reset_index(name='HOST_VERIFICATION_ID')
```

```python
tab = tab.rename(index=str, columns={"host_id": "HOST_ID"})
tab
```
