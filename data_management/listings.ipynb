```python
import numpy as np
import pandas as pd
```

## Listings

```python
barcelona_listings = pd.read_csv("barcelona_listings.csv")
berlin_listings = pd.read_csv("berlin_listings_filtered.csv")
madrid_listings = pd.read_csv("madrid_listings_filtered.csv")
```

```python
#check which neighboorhood to use
listing_cols = ['id','neighbourhood','host_id','cancellation_policy','bed_type','room_type',
       'property_type','longitude','latitude','summary','name', 'listing_url','interaction',
       'access','notes','neighborhood_overview','description','space','transit',
       'house_rules', 'picture_url','accommodates','bathrooms','bedrooms','beds','square_feet',
       'price','guests_included','weekly_price','monthly_price','security_deposit',
       'cleaning_fee','extra_people','minimum_nights','maximum_nights','review_scores_accuracy',
       'review_scores_cleanliness','review_scores_checkin','review_scores_communication',
       'review_scores_location','review_scores_value','review_scores_rating',
               'require_guest_profile_picture', 'require_guest_phone_verification',
               'is_business_travel_ready']
```

```python
berlin_listings.columns
```

```python
all_listings = pd.concat([berlin_listings,
                           madrid_listings,
                           barcelona_listings])
listings = all_listings[listing_cols]
```

```python
neighs = pd.read_csv('neighbourhood.csv')
cancel_pol = pd.read_csv('cancel_pol.csv')
bed_type = pd.read_csv('bed_types.csv')
room_type = pd.read_csv('room_type.csv')
property_type = pd.read_csv('property_type.csv')
```

```python
#all_listings['property_type'].head(30)
```

```python
#property_type.head(30)
```

```python
#MAP NEIGHBORHOOD TO ID
neigh_dict = pd.Series(neighs['NEIGH_ID'].values,index=neighs['NEIGH']).to_dict()
listings = listings.replace({"neighbourhood": neigh_dict})
#MAP CANCELLATION POLICY TO ID
cancel_dict = pd.Series(cancel_pol['CANCELLATION_POLICY_ID'].values,
                        index=cancel_pol['CANCELLATION_POLICY']).to_dict()
listings = listings.replace({"cancellation_policy": cancel_dict})
#MAP BED TYPE TO ID
bed_dict = pd.Series(bed_type['BED_TYPE_ID'].values,
                        index=bed_type['BED_TYPE']).to_dict()
listings = listings.replace({"bed_type": bed_dict})
#MAP ROOM TYPE TO ID
room_dict = pd.Series(room_type['ROOM_TYPE_ID'].values,
                        index=room_type['ROOM_TYPE']).to_dict()
listings = listings.replace({"room_type": room_dict})
#MAP PROPERTY TYPE TO ID
property_dict = pd.Series(property_type['PROPERTY_TYPE_ID'].values,
                        index=property_type['PROPERTY_TYPE']).to_dict()
listings = listings.replace({"property_type": property_dict})

#all_listings['property_type'].head(30)
listings = listings.rename(columns={'access': 'listing_access', 'neighbourhood': 'neigh_id',
                                   'cancellation_policy':'cancellation_policy_id',
                                   'bed_type':'bed_type_id','room_type':'room_type_id',
                                   'property_type':'property_type_id','id': 'listing_id'})
listings = listings.set_index('listing_id')
```

```python
#listings.to_csv('listings.csv')
```

```python
listings['price'] = listings['price'].str.replace('$','').str.replace(',','').astype(float)
listings['weekly_price'] = listings['weekly_price'].str.replace('$','').str.replace(',','').astype(float)
listings['monthly_price'] = listings['monthly_price'].str.replace('$','').str.replace(',','').astype(float)
listings['security_deposit'] = listings['security_deposit'].str.replace('$','').str.replace(',','').astype(float)
listings['cleaning_fee'] = listings['cleaning_fee'].str.replace('$','').str.replace(',','').astype(float)
listings['extra_people'] = listings['extra_people'].str.replace('$','').str.replace(',','').astype(float)
listings.to_csv('listings.csv')
listings.info()
```
